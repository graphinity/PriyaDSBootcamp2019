```python
Dataframes = []

for filename in list_of_files:
    Dataframes.append(pd.read_csv(filename))
```

```python
Masterdf= pd.concat(Dataframes, axis=0)
```

```python
Masterdf.shape
```

```python
Masterdf.info()
```

```python
Masterdf['Date/Time']=pd.to_datetime(Masterdf['Date/Time'])
```

```python
Masterdf.info()
```

```python
Masterdf['Month']= Masterdf['Date/Time'].dt.month
```

```python
Masterdf['Day']= Masterdf['Date/Time'].dt.day
```

```python
Masterdf['dayofweek']=Masterdf['Date/Time'].dt.dayofweek
```

```python
Masterdf['hour']=Masterdf['Date/Time'].dt.hour
```
